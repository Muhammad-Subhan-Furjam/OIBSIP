```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C://Users//subha//Downloads//archive (1)//Unemployment_data.csv")
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='Date', y='Estimated Unemployment Rate (%)', data=df)
plt.title('Unemployment Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Compare regions during Covid-19
covid_period_start = '2020-03-01'
covid_period_end = '2021-02-28'
```

```python
covid_data = df[(df['Date'] >= covid_period_start) & (df['Date'] <= covid_period_end)]
```

```python
if not covid_data.empty:
    if 'Region' in covid_data.columns:
        covid_data = covid_data.dropna(subset=['Region'])
        if not covid_data['Region'].empty:
            plt.figure(figsize=(10, 6))
            sns.boxplot(x='Region', y='Estimated Unemployment Rate (%)', data=covid_data)
            plt.title('Unemployment Rate Distribution During Covid-19')
            plt.xlabel('Region')
            plt.ylabel('Estimated Unemployment Rate (%)')
            plt.xticks(rotation=45)
            plt.show()
        else:
            print("No valid data points found for the specified region during the Covid-19 period.")
    else:
        print("'Region' column not found in the dataset.")
else:
    print("No data found for the specified Covid-19 period.")
```
